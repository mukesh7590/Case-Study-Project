```python
# Quetion 2

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv("/startup_funding.csv")

#Cleaning InvestmentType
hash=dict()

for string in df['InvestorsName']:
    string=str(string)
    word= string.split(',')
    for ch in word:
        ch=ch.strip()
        if ch in hash:
            hash[ch]+=1
        else:
            hash[ch]=1
            
X=list()
Y=list()

hash = sorted(hash.items(), key=lambda x: x[1], reverse=True)[0:5]
for ele in hash:
    X.append(ele[0])
    Y.append(ele[1])
    print(ele[0],end=": ")
    print(ele[1])
    
plt.title('Top 5 Investors')
plt.pie(Y, labels = X, autopct = "%.2f%%") 
plt.axis("equal")
plt.show()

```

```python
# answer 3 final 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv('/startup_funding.csv',delimiter=',',skipinitialspace=True)
df['StartupName'].dropna(inplace=True)
df['StartupName'].replace('Ola Cabs','Ola',inplace=True)
df['StartupName'].replace('Olacabs','Ola',inplace=True)
df['StartupName'].replace('Flipkart.com','Flipkart',inplace=True)
df['StartupName'].replace('Paytm Marketplace','Paytm',inplace=True)
df['StartupName'].replace('Oyo Rooms','Oyo',inplace=True)
df['StartupName'].replace('Oyorooms','Oyo',inplace=True)
df['StartupName'].replace('OyoRooms','Oyo',inplace=True)

hash=dict()
data=list(zip(df['InvestorsName'],df['StartupName']))

def addValue(hash,company,investor):
    if company in hash[investor]:
        return True
    return False

for ele in data:
    str1 = str(ele[0])
    str2 = str(ele[1])
    words = str1.split(',')
    company = str2.strip()
    for investor in words:
        investor = investor.strip()
        if investor != '':
            if investor in hash:
                if not(addValue(hash,company,investor)):
                    hash[investor].append(company)
            else:
                arr=[]
                hash[investor]=arr
                hash[investor].append(company)
                
for key in hash:
    hash[key]=len(hash[key])
hash = sorted(hash.items(), key=lambda x: x[1], reverse=True)[0:5]
X=[]
Y=[]
for ele in hash:
    X.append(ele[0])
    Y.append(ele[1])
    print(ele[0],end=": ")
    print(ele[1])
plt.pie(Y, labels = X, autopct = "%.2f%%") 
plt.title('Top 5 Investors who invest in different StartUp')
plt.axis("equal")
plt.show()
```
